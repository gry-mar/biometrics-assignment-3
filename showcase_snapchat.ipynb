```python
errors = []
import glob
from tqdm import tqdm
import os
import random

RECREATE_SNAPCHAT = False

if RECREATE_SNAPCHAT:
    random.seed(42222)
    bar = tqdm(glob.glob("dataset/**/*.jpeg", recursive=True))
    for img in bar:
        try:
            operation = random.choice([add_flowers, add_lips, add_sunglasses])
            transformed = operation(img)
            os.makedirs(
                os.path.dirname(img).replace("dataset", "snapchat"), exist_ok=True
            )
            plt.imsave(
                os.path.dirname(img).replace("dataset", "snapchat")
                + f"/{os.path.basename(img)}",
                transformed,
            )
        except Exception as e:
            errors.append((img, e))
```
