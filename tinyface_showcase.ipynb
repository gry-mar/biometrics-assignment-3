```python
import matplotlib.pyplot as plt
import glob
import random
import cv2
```

```python
faces = glob.glob("tinyface/Testing_Set/Gallery_Match/*.jpg")
random.seed(13445)

sample = random.sample(faces, 100)
plt.figure(figsize=(20, 20))
for i, face in enumerate(sample):
    plt.subplot(10, 10, i + 1)
    image = plt.imread(face)
    image = cv2.resize(
        image,
        (2 * image.shape[1], 2 * image.shape[0]),
        interpolation=cv2.INTER_NEAREST_EXACT,
    )
    plt.imshow(image)
    plt.axis("off")
```
